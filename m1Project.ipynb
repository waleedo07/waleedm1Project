```python
import numpy as np
import pandas as pd
```

```python
df_carSale = pd.read_csv('D:\Study\semester 5\data engineering and visualization\projects and assignments\Car_sale_ads.csv')
```

```python
df_carSale.head(10)
```

```python
def missingPercent(x):
 percent_missing = x.isnull().sum() * 100 / len(x)
 missing_value_df = pd.DataFrame({'column_name': x.columns,
                                 'percent_missing': percent_missing})
 return missing_value_df
```

```python
print(missingPercent(df_carSale))
```

```python
df_carSale.pop('First_owner')
df_carSale.pop('First_registration_date')
df_carSale.pop('CO2_emissions')
df_carSale.pop('Vehicle_version')
df_carSale.pop('Vehicle_generation')
df_carSale.pop('Origin_country')
df_carSale = df_carSale[df_carSale['Mileage_km'].notna()]
df_carSale = df_carSale[df_carSale['Power_HP'].notna()]
df_carSale = df_carSale[df_carSale['Displacement_cm3'].notna()]
df_carSale = df_carSale[df_carSale['Transmission'].notna()]
df_carSale = df_carSale[df_carSale['Doors_number'].notna()]
df_carSale.replace('[]', np.nan, inplace=True)
df_carSale.replace('other', np.nan, inplace=True)
```

```python
print(missingPercent(df_carSale))
```

```python
print(sum(df_carSale['Currency'] == 'PLN'))
print(sum(df_carSale['Currency'] =='EUR'))
```

```python
df_carSale = df_carSale[df_carSale.Currency !='EUR']
```

```python
df_carSale.pop('Currency')
```

```python
df_carSale['Drive']=df_carSale['Drive'].fillna(df_carSale['Drive'].mode()[0])
df_carSale['Colour']=df_carSale['Colour'].fillna(df_carSale['Colour'].mode()[0])
df_carSale['Features']=df_carSale['Features'].fillna(df_carSale['Features'].mode()[0])
```

```python
print(missingPercent(df_carSale))
df_carSale.head(10)
```

```python
from sklearn.datasets import load_boston
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
z = np.abs(stats.zscore(df_carSale['Price']))
filtered_entries = z < 3
df_zscore_filter = df_carSale[filtered_entries]

df_carSale=df_zscore_filter
```

```python
z = np.abs(stats.zscore(df_carSale['Mileage_km']))
filtered_entries = z < 3
df_zscore_filter = df_carSale[filtered_entries]
print(df_zscore_filter.shape)
print(df_carSale.shape)
df_carSale=df_zscore_filter
```

```python
z = np.abs(stats.zscore(df_carSale['Power_HP']))
filtered_entries = z < 3
df_zscore_filter = df_carSale[filtered_entries]
print(df_zscore_filter.shape)
print(df_carSale.shape)
df_carSale=df_zscore_filter
```
